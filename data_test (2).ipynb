```python
import numpy as np
import random
import pandas as pd
```

```python
cdf = pd.read_csv('Crop_Data.csv')
```

```python
# List of all distinct crops
crops = list(cdf.iloc[:,0])
print(crops)
```

```python
# define number of entries for each crop
n_entries = 150
```

```python
data = cdf.values.tolist()

# gen_data will hold all the generated values
gen_data = [[0 for columns in range(5)] for rows in range(n_entries*len(crops))]
```

```python
# This function will add 'n_entries' number of entries of each crop

def add_entry(i):
    for j in range(n_entries):
      # gen_data[i*100 + j][0] = np.random.uniform(data[i][1], data[i][2])    # N
      # gen_data[i*100 + j][1] = np.random.uniform(data[i][3], data[i][4])    # P
      # gen_data[i*100 + j][2] = np.random.uniform(data[i][5], data[i][6])    # K
      gen_data[i*n_entries + j][0] = np.random.uniform(data[i][7], data[i][8])    # Temp
      gen_data[i*n_entries + j][1] = np.random.uniform(data[i][9], data[i][10])   # Hum
      gen_data[i*n_entries + j][2] = np.random.uniform(data[i][11], data[i][12])  # pH
      gen_data[i*n_entries + j][3] = np.random.uniform(data[i][13], data[i][14])  # Rain
      gen_data[i*n_entries + j][4] = data[i][0]
```

```python
for i in range(len(crops)):
  add_entry(i)
```

```python
# df dataframe contains the final generated dataframe which we will copy into a csv file

df = pd.DataFrame(gen_data, columns = ['Temp', 'Hum', 'pH', 'Rain', 'Crop'])
df.to_csv('gen_dataset.csv', index=False)
```
